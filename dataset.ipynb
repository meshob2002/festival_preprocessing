# API 통한 데이터셋 추출

 - 기본 축제 데이터(이름, 설명) -> festival_ds.csv 데이터셋

```python
# 라이브러리 import
import requests
import pprint
import json

# url 입력 지역축제 api 활용

url = 'http://api.kcisa.kr/openapi/service/rest/meta4/getKCPG0504?serviceKey=e9ba5951-a431-4615-ae38-1e947f0a4e3b&numOfRows=11356&pageNo=1'

# url 불러오기
response = requests.get(url)

#데이터 값 출력해보기
contents = response.text
```

```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd
# XML 파싱
root = ET.fromstring(contents)

# 필요한 필드 추출하여 딕셔너리 생성
items = root.findall('body/items/item')  # item 요소들을 가져옴
data_dict = {
    'title': [],
    'alternativeTitle': [],
    'creator': [],
    'regDate': [],
    'collectionDb': [],
    'subjectCategory': [],
    'subjectKeyword': [],
    'extent': [],
    'description': [],
    'spatialCoverage': [],
    'temporalCoverage': [],
    'person': [],
    'language': [],
    'sourceTitle': [],
    'referenceIdentifier': [],
    'rights': [],
    'copyrightOthers': [],
    'url': [],
    'contributor': []
}

for item in items:
    data_dict['title'].append(item.find('title').text)
    data_dict['alternativeTitle'].append(item.find('alternativeTitle').text)
    data_dict['creator'].append(item.find('creator').text)
    data_dict['regDate'].append(item.find('regDate').text)
    data_dict['collectionDb'].append(item.find('collectionDb').text)
    data_dict['subjectCategory'].append(item.find('subjectCategory').text)
    data_dict['subjectKeyword'].append(item.find('subjectKeyword').text)
    data_dict['extent'].append(item.find('extent').text)
    data_dict['description'].append(item.find('description').text)
    data_dict['spatialCoverage'].append(item.find('spatialCoverage').text)
    data_dict['temporalCoverage'].append(item.find('temporalCoverage').text)
    data_dict['person'].append(item.find('person').text)
    data_dict['language'].append(item.find('language').text)
    data_dict['sourceTitle'].append(item.find('sourceTitle').text)
    data_dict['referenceIdentifier'].append(item.find('referenceIdentifier').text)
    data_dict['rights'].append(item.find('rights').text)
    data_dict['copyrightOthers'].append(item.find('copyrightOthers').text)
    data_dict['url'].append(item.find('url').text)
    data_dict['contributor'].append(item.find('contributor').text)

# 데이터 프레임 생성
df = pd.DataFrame.from_dict(data_dict)


# 데이터 프레임 생성
df = pd.DataFrame.from_dict(data_dict)

# 결과 확인
df.to_csv('festival_ds.csv', index=False)
```

```python
#사용 가능한 컬럼만 남김
df = df.drop(['alternativeTitle','creator','regDate','collectionDb','extent','spatialCoverage','temporalCoverage','person','language','rights','contributor'], axis=1)
```

# 트렌드(검색량 데이터셋)
 - 2018~2022년 데이터셋 합침 -> trend.csv 데이터
 - 각데이터셋의 컬럼 정의서는 엑셀파일로 첨부 되어있습니다
 - 사용할려면 축제기간별 검색량 데이터가 합쳐져있음(ex 제27회 화도진축제 ,제28회 화도진축제의 검색량이 같음 하나의 데이터로 봐야함-> 화도진축제(27,28~~) 중복값 처리가 필요함 정규식 써서 해야됨(나중에)) 

```python
trend1 = pd.read_csv("KC_594_WNTY_FSTV_TRND_MAP_2019.csv")
# TIME 열 생성
trend1['TIME'] = trend1.groupby('FCLTY_NM')['SEARCH_MT'].transform(lambda x: x.min() + '~' + x.max())
trend1['SEARCH_SUM'] = trend1.groupby('FCLTY_NM')['SEARCH_CO'].transform('sum')
trend1['RANK_AVG'] = trend1.groupby('FCLTY_NM')['RANK_CO'].transform('mean')


# 불필요한 열 삭제
trend1 = trend1.drop(['FCLTY_ID','SEARCH_MT','SEARCH_CO','RANK_CO','LEGALDONG_CD','ADSTRD_CD','RDNMADR_CD','FILE_NM','BASE_DE'], axis=1).drop_duplicates()
# 중복된 행 제거
trend1 = trend1.drop_duplicates().reset_index(drop=True)
# 결과 출력
trend1
```

```python
trend4 = pd.read_csv("KC_594_WNTY_FSTV_TRND_MAP_2022.csv") # 위도 경도 컬럼명 바뀜
trend4['SEARCH_MT'] = pd.to_datetime(trend4['SEARCH_MT'], format='%y-%b').dt.strftime('%Y-%m')
# TIME 열 생성
trend4['TIME'] = trend4.groupby('FCLTY_NM')['SEARCH_MT'].transform(lambda x: x.min() + '~' + x.max())
trend4['SEARCH_SUM'] = trend4.groupby('FCLTY_NM')['SEARCH_CO'].transform('sum')
trend4['RANK_AVG'] = trend4.groupby('FCLTY_NM')['RANK_CO'].transform('mean')


# 불필요한 열 삭제
trend4 = trend4.drop(['FCLTY_ID','SEARCH_MT','SEARCH_CO','RANK_CO','LEGALDONG_CD','ADSTRD_CD','RDNMADR_CD','FILE_NM','BASE_DE'], axis=1).drop_duplicates()
# 중복된 행 제거
trend4 = trend4.drop_duplicates().reset_index(drop=True)
# 결과 출력
trend4
```

```python
trend_concat = pd.concat([trend1, trend2, trend3, trend4]).reset_index(drop=True)
#트렌드 데이터셋의 경우 중복되는 값 처리필요 ex(흥부제가 2018,2019,2020 나뉘어져있음 + 2018축제이름2019축제이름2020축제이름 이러한 데이터셋도 처리 필요)
trend_concat.to_csv('trend.csv',index=False)
```

# 축제 및 주변관광지 데이터셋
  - 2017~2022년 데이터셋 합침 -> near.csv 데이터셋
  - 각 컬럼정의서는 엑셀파일에 첨부되어있습니다
  - 2022년 데이터셋과 2023년 데이터셋의 TRRSRT_CL_NM 관광지 분류 정보가 2021년 데이터셋과 다르게 분류 코드형식으로 되어 있음
  

```python
near1 = pd.read_csv('KC_CLTUR_FSTVL_NEARBY_TURSM_INFO_2020.csv')
near1 = near1[['FCLTY_NM','CTPRVN_NM','SIGNGU_NM','LEGALDONG_NM','ADSTRD_NM','RN_ADDR','LA','LO','OPAR_NM','FSTVL_BEGIN_DE','FSTVL_END_DE','FSTV_CN','RELATE_INFO_CN','TRRSRT_POI_NM','TRRSRT_CL_NM','TRRSRT_LA','TRRSRT_LO']]
```

```python
# 주변 관광지 이름이 같지만 좌표가 다른 데이터가 존재하여 중복제거를 위한 전처리 작업
near1[['TRRSRT_LA', 'TRRSRT_LO']] = near1.groupby('TRRSRT_POI_NM')[['TRRSRT_LA', 'TRRSRT_LO']].transform('first')
```

```python
# FCLTY_NM 기준으로 TRRSRT_POI_NM, TRRSRT_CL_NM, TRRSRT_LA, TRRSRT_LO의 값을 리스트로 추출하여 저장
near1_grouped = near1.groupby('FCLTY_NM').agg({'TRRSRT_POI_NM': lambda x: list(x),
                                         'TRRSRT_CL_NM': lambda x: list(x),
                                         'TRRSRT_LA': lambda x: list(x),
                                         'TRRSRT_LO': lambda x: list(x)}).reset_index()
```

```python
# Merge with the original dataset
near1_merged = pd.merge(near1, near1_grouped, on='FCLTY_NM', how='left')
```

```python
near1_merged =near1_merged.drop(['TRRSRT_POI_NM_x','TRRSRT_CL_NM_x','TRRSRT_LA_x','TRRSRT_LO_x'],axis=1).drop_duplicates(subset=['FCLTY_NM']).reset_index(drop=True)
```

```python
near2 = pd.read_csv('KC_CLTUR_FSTVL_NEARBY_TURSM_INFO_2021.csv')
near2 = near2[['FCLTY_NM','CTPRVN_NM','SIGNGU_NM','LEGALDONG_NM','ADSTRD_NM','FCLTY_ROAD_NM_ADDR','FCLTY_LA','FCLTY_LO','OPAR_NM','FSTVL_BEGIN_DE','FSTVL_END_DE','FSTVL_CN','RELATE_INFO_CN','TRRSRT_POI_NM','TRRSRT_CL_NM','TRRSRT_LA','TRRSRT_LO']]
# FCLTY_NM 기준으로 TRRSRT_POI_NM, TRRSRT_CL_NM, TRRSRT_LA, TRRSRT_LO의 값을 리스트로 추출하여 저장
# 2020년 데이터셋의 컬럼값일치 X 변경후 진행
near2 = near2.rename(columns={'FCLTY_ROAD_NM_ADDR': 'RN_ADDR', 'FCLTY_LA': 'LA', 'FCLTY_LO': 'LO', 'FSTVL_CN': 'FSTV_CN'})
near2_grouped = near2.groupby('FCLTY_NM').agg({'TRRSRT_POI_NM': lambda x: list(x),
                                         'TRRSRT_CL_NM': lambda x: list(x),
                                         'TRRSRT_LA': lambda x: list(x),
                                         'TRRSRT_LO': lambda x: list(x)}).reset_index()
# Merge with the original dataset
near2_merged = pd.merge(near2, near2_grouped, on='FCLTY_NM', how='left')
near2_merged =near2_merged.drop(['TRRSRT_POI_NM_x','TRRSRT_CL_NM_x','TRRSRT_LA_x','TRRSRT_LO_x'],axis=1).drop_duplicates(subset=['FCLTY_NM']).reset_index(drop=True)
near2_merged
```

```python
near3 = pd.read_csv('KC_CLTUR_FSTVL_NEARBY_TURSM_INFO_2022.csv')
near3 = near3[['FCLTY_NM','CTPRVN_NM','SIGNGU_NM','LEGALDONG_NM','ADSTRD_NM','FCLTY_ROAD_NM_ADDR','FCLTY_LA','FCLTY_LO','OPAR_NM','FSTVL_BEGIN_DE','FSTVL_END_DE','FSTVL_CN','RELATE_INFO_CN','TRRSRT_POI_NM','TRRSRT_CL_NM','TRRSRT_LA','TRRSRT_LO']]
# FCLTY_NM 기준으로 TRRSRT_POI_NM, TRRSRT_CL_NM, TRRSRT_LA, TRRSRT_LO의 값을 리스트로 추출하여 저장
# 2020년 데이터셋의 컬럼값일치 X 변경후 진행
near3 = near3.rename(columns={'FCLTY_ROAD_NM_ADDR': 'RN_ADDR', 'FCLTY_LA': 'LA', 'FCLTY_LO': 'LO', 'FSTVL_CN': 'FSTV_CN'})
near3_grouped = near3.groupby('FCLTY_NM').agg({'TRRSRT_POI_NM': lambda x: list(x),
                                         'TRRSRT_CL_NM': lambda x: list(x),
                                         'TRRSRT_LA': lambda x: list(x),
                                         'TRRSRT_LO': lambda x: list(x)}).reset_index()
# Merge with the original dataset
near3_merged = pd.merge(near3, near3_grouped, on='FCLTY_NM', how='left')
near3_merged =near3_merged.drop(['TRRSRT_POI_NM_x','TRRSRT_CL_NM_x','TRRSRT_LA_x','TRRSRT_LO_x'],axis=1).drop_duplicates(subset=['FCLTY_NM']).reset_index(drop=True)
near3_merged
```

```python
near_concat = pd.concat([near1_merged, near2_merged, near3_merged]).reset_index(drop=True)
#주변관광지 데이터셋의 경우도 마찬가지로 중복되는 값 처리필요 ex(흥부제가 2018,2019,2020 나뉘어져있음 + 2018축제이름2019축제이름2020축제이름 이러한 데이터셋도 처리 필요)
near_concat.to_csv('near.csv',index=False)
near_concat
```
